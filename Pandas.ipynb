```python
import pandas as pd

vendas = {
    'data': ['15/02/2021', '16/02/2021'],
    'valor': [100, 200],
    'produto': ['camiseta', 'camiseta'],
    'quantidade': [2, 3],
}

vendas_df = pd.DataFrame(vendas)

display(vendas_df)
```

```python
vendas_df = pd.read_excel("Vendas.xlsx")
display(vendas_df)
```

```python
#.head -> mostra apenas a cabeça do arquivo
display(vendas_df.head)
```

```python
#.head(x) -. x mostra a quantidade de linhas que quer visualizar do arquivo
display(vendas_df.head(10))
```

```python
#.shape -> mostra a quantidade de linhas e de colunas 
display(vendas_df.shape)
```

```python
#.describe() -> calcula a média, da o desvio padrão, da o minímo, o máximo, um quartio. Visão geral das informações
display(vendas_df.describe()) 
```

```python
#Pegar 1 Coluna -> Apenas criar uma variável e passar como paramêtro o nome da coluna.
produto = vendas_df['Produto']
display(produto)
```

```python
#Pegar 2 Colunas -> mesmo método que pegar 1 coluna, apenas adicionar o nome da coluna que quer pegar, apenas adicionar a mais []
produtos = vendas_df[['Produto', 'ID Loja']]
display(produtos)
```

```python
#.loc -> pegar uma linha, ele localiza as linhas pelos seus IDS 
display(vendas_df.loc[1])
```

```python
#.loc -> pegar uma sequência de linhas, ele localiza as linhas pelos seus IDS 
display(vendas_df.loc[1:5])
```

```python
#.loc -> pegar linhas que correspondem a uma condição, passar dentro do loc
#coluna_loja -> nome_loja
#vendas_df['ID Loja'] == 'Norte Shopping'
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']
display(vendas_norteshopping_df)
```

```python
#pegar várias linhas e colunas usando o loc
#vendas_norteshopping_df = vendas_df.loc[linhas, colunas]
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]
display(vendas_norteshopping_df)
```

```python
#.loc -> Pegar um valor específico  
#display(vendas_df.loc[linha, coluna])
display(vendas_df.loc[1, 'Produto'])
```

```python
#ADICIONAR 1 COLUNA
#A partir de uma coluna 
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)
```

```python
#ADICIONAR 1 COLUNA
#Criar uma coluna com valor padrão
#vendas_df.loc[linha, coluna] = 0
#vendas_df.loc[:, "Imposto"]] = 0
vendas_df.loc[:, "Imposto"] = 0
display(vendas_df) 
```

```python
vendas_dez_df = pd.read_excel("Vendas - Dez.xlsx")
display(vendas_dez_df)
```

```python
#ADICIONAAR 1 LINHA 
#Linha de um complemento de base de dados 
vendas_df = vendas_df.append(vendas_dez_df)
display(vendas_df)
```

```python
#EXCLUIR LINHAS E COLUNAS 
vendas_df = vendas_df.drop("Imposto", axis=1)
display(vendas_df)
```

```python
 # excluir determinada linha
 #EIXO (AXIS) 0 É LINHA , E EIXO (AXIS) 1 É COLUNA
 vendas_df = vendas_df.drop(0, axis=0)
 display(vendas_df)
```

```python
# Deletar linhas e colunas completamentes vazias
vendas_df = vendas_df.dropna(how = 'all')
display(vendas_df)
```

```python
# Deletar linhas completamentes vazias
vendas_df = vendas_df.dropna(how='all', axis=0)
```

```python
#Deletar colunas completamente vazias
vendas_df = vendas_df.dropna(how='all', axis=1)
```
